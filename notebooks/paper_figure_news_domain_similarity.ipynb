```python
import pandas as pd
import matplotlib.pyplot as plt
import scicolor

import seaborn as sns

plt.rcParams.update({"font.size": 14})
plt.rcParams.update({"axes.axisbelow": True})
```

```python
similarity_matrix_cosine = pd.read_parquet(
    "../data/intermediate/citation_analysis/news_similarity_matrix_cosine.parquet"
)
```

```python
models = [
    "gpt-4o-mini-search-preview",
    "gpt-4o-search-preview",
    "gpt-4o-search-preview-high",
    "gpt-4o-search-preview-high-loc",
    "sonar",
    "sonar-pro",
    "sonar-pro-high",
    "sonar-reasoning",
    "sonar-reasoning-pro-high",
    "gemini-2.0-flash-grounding",
    "gemini-2.5-flash-preview-04-17-grounding",
    "gemini-2.5-pro-exp-03-25-grounding",
]
```

```python
similarity_matrix_cosine.columns
```

```python
plt.figure(figsize=(10, 8))
ax = sns.heatmap(
    similarity_matrix_cosine[models].loc[models],
    vmin=0,
    vmax=1,
    cmap=scicolor.get_cmap("acton"),
    square=True,
)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig(
    "../data/output/cleaned_arena_data/model_news_domain_cosine_similarity_heatmap.pdf"
)
```
